```python
# Import dependency

import pandas as pd

df = pd.read_csv("../Resources/reviews.csv")
df.head()
```

```python
# Remove columns of limited/no value to analysis
df.drop(['brand','author','date','title','taste','ingredients','texture','likes'], axis=1, inplace=True)
```

```python
# Remove special characters that will cause issues in postgres import
df['text'] = df['text'].replace({'\n':' '},regex=True)
df['text'] = df['text'].replace({'\'':''},regex=True)
```

```python
# Create helpful reviews table
df_helpful = df[df['helpful_yes'] > df['helpful_no']]
```

```python
# Send both tables back out as CSVs
df_helpful.to_csv("../Resources/helpful_clean_reviews.csv", index=False)
df.to_csv("../Resources/clean_reviews.csv", index=False)
```
