```python
import pandas as pd
```

```python
df_tokenized = pd.read_csv("Resources/engineered_feature_tokens.csv")
#drop index 
df_tokenized = df_tokenized.drop(["Unnamed: 0"], axis=1)
df_tokenized.head(2)
```

```python
df_vader = pd.read_csv("Resources/VADER_sentiment_score.csv")
df_vader = df_vader.drop(["Unnamed: 0"], axis=1)
df_vader.head(2)
```

```python
compound_merge = df_vader.merge(df_tokenized,left_index=True, right_index=True)
compound_merge = compound_merge.drop(["key_y","stars_y","helpful_yes","helpful_no","rating","sentiment_y"], axis=1)
compound_merge.head(2)
```

```python
# Rename the columns
compound_merge.rename(columns = {'key_x':'key', 'text_x':'text','stars_x':'stars','sentiment_x':'sentiment',
                                'text_y':'tokens'}, inplace = True)
compound_merge.head(2)
```

```python
compound_merge.to_csv("Resources/compound_merge.csv")
```
